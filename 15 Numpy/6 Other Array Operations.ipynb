```python
np.repeat(a, 3) #output in 1d again for a 3d array too, repeat gives value after flattening the array.
```

```python
a
```

```python
b=np.repeat(a, 3, axis=0) #row wise repetition, same dimensions
print(b.ndim)
print(b)
```

#### np.diag()
Extract a diagonal or construct a diagonal array
<br> <br>
  If v is a 2-D array, return a copy of its k-th diagonal. If v is a 1-D array, return a 2-D array with v on the k-th diagonal.
k : int, optional
    Diagonal in question. The default is 0. Use k>0 for diagonals above the main diagonal, and k<0 for diagonals below the main diagonal   

```python
np.diag(np.array([1,2,3,4]))
```

```python
arr=np.random.randint(1,10, size=(3,3))
arr
```

```python
arr.diagonal()
```

```python
np.diag(arr, k=0)
```

```python
np.diag(arr, k=1)
```

```python
np.diag(arr, k=2)
```

```python
np.diag(arr, k=-1)
```

```python
np.diag(arr, k=-2)
```
